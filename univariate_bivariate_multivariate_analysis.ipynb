```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# this is most useful and basic step for exploratory data analysis 
#in univariate we use single feature to analyize out o/p
# in bi variate we use 2 feature in x and y plot 
# in multivariate we basically use seaborn lib and pair plot as we can not ble to view more than 3d data 
```

```python
df = pd.read_csv('iris.csv')
```

```python
df
```

# univariate analysis

```python
df_setosa = df.loc[df['Species']=='Iris-setosa']
df_virginica = df.loc[df['Species']=='Iris-virginica']
df_versicolor = df.loc[df['Species']=='Iris-versicolor']
```

```python
# here we are using only one feature randomly just take sepallength
```

```python
plt.plot(df_setosa['SepalLengthCm'],np.zeros_like(df_setosa['SepalLengthCm']),'or')
plt.plot(df_virginica['SepalLengthCm'],np.zeros_like(df_virginica['SepalLengthCm']),'+g')
plt.plot(df_versicolor['SepalLengthCm'],np.zeros_like(df_versicolor['SepalLengthCm']))
plt.show()
```

```python
# here we can see lots of overlapping so we should not use this
```

# Bivariate analysis

```python
sns.FacetGrid(df,hue='Species',size=6).map(plt.scatter,'PetalLengthCm','SepalWidthCm')
plt.legend()
plt.show()
```

```python
# here also we can see overlapping b/w versicolor and virginica
```

# multivariate analysis 

```python
# when more than 2 or 3 feature we use this 
```
