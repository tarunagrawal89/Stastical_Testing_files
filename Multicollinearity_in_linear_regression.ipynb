```python
import pandas as pd 
import statsmodels.api as sm
```

```python
# now we see multicollinearity in linear regression model for this we need to know about correlation first
```

```python
df = pd.read_csv('advertising.csv')
```

```python
df
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
 # in this case we have equation like sale = b0 + b1*tv + b2*radio + b3*newspaper so we adding b0 constant here
```

```python
x = sm.add_constant(x)
```

```python
x
```

```python
#now fit a ols model - ordinary least square
model = sm.OLS(y,x).fit()
```

```python
model.summary()
# here r squared is very good and most important std error is very less if there would be any correlation
# b/w any feature then this std error will be more so there is no multicollinearity in this dataset.  
```

```python
# now there is one another way to find out correlation b/w features by using corr()
```

```python
import matplotlib.pyplot as plt
```

```python
x.corr()
```

```python
# here also we can see there is very less nearre to zero correlation b/w each independent feature 
```
